```python
from pdf2image import convert_from_path
import cv2
import numpy as np
# Use a rectangular kernel for morphological operations
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (1, 1))


# Function to perform OCR on a PDF file
def ocr_pdf(pdf_path):
    # Convert PDF to a list of images
    pages = convert_from_path(pdf_path, 300)  # 300 is the DPI (resolution)

    # Initialize an empty string to store text
    extracted_text = ""

    config = r'--oem 3 --psm 3'

    # Perform OCR on each image
    for page_number, image in enumerate(pages):
        gray = cv2.cvtColor(np.array(image), cv2.COLOR_BGR2GRAY)

        # Apply Otsu's thresholding
        _, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

        # Apply morphological opening (erosion followed by dilation)
        opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel)

        # Dilate to connect text
        dilated = cv2.dilate(opening, kernel, iterations=1)

        text = pytesseract.image_to_string(image, lang='vie',config=config) # specify language here
        extracted_text += f"--- Page {page_number+1} ---\n{text}\n"

    return extracted_text

# IMPORTANT: ONLY CHANGE HERE (Path to your PDF file)
pdf_path = '/content/04-ttg.signed.pdf' #change the PDF file name here

# Perform OCR
extracted_text = ocr_pdf(pdf_path)
print(extracted_text)
```

Output the file

```python

# Specify the file path and mode (w for write, a for append)
file_path = 'output.txt'

# Open the file in write mode
with open(file_path, 'w') as file:
    # Write the string to the file
    file.write(extracted_text)


```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("VietAI/vit5-base")
model = AutoModelForSeq2SeqLM.from_pretrained("VietAI/vit5-base")
```

```python
sentences = extracted_text.split('\n')
sentences[11]
```

```python
model.to("cuda")
src = """
### Input:
Căn cứ Luật Tổ chức Chính phú ngày 19 tháng 6 năm 2015; Luật sửa đổi, bỗ sung một số điều của Luật Tổ chức Chính phủ và Luật TỔ chức chính quyền địa
phương ngày 22 tháng I1 năm 2019,
### Response:
Căn cứ Luật Tổ chức Chính phú ngày 19 tháng 6 năm 2015; Luật sửa đổi, bổ sung một số điều của Luật Tổ chức Chính phủ và Luật TỔ chức chính quyền địa
phương ngày 22 tháng 11 năm 2019,
### Input:
"""

tokenized_text = tokenizer.encode(src, return_tensors="pt").to("cuda")
model.eval()
summary_ids = model.generate(
                    tokenized_text,
                    max_length=256,
                    num_beams=5,
                    repetition_penalty=2.5,
                    length_penalty=1.0,
                    early_stopping=True
                )
output = tokenizer.decode(summary_ids[0], skip_special_tokens=False)
print(output)
```
